```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline


image = mpimg.imread('human/b1.jpg')

plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
sobel_y = np.array([[ -1,2, -1], 
                   [ -1, 5, -1], 
                   [ -1, 2, -1]])


sobel_x = np.array([[ 1,8, 1], 
                   [ 2, 8, 2], 
                   [ 1, 8, 1]])
  
filtered_image = cv2.filter2D(gray, -1, sobel_y)

plt.imshow(filtered_image, cmap='gray')
```

```python

get_ipython().magic('matplotlib inline')
test_image = cv2.imread('human/b1.jpg')
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
plt.imshow(test_image_gray, cmap='gray')
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('c:/users/LEGION/projetf/haaracascade/haarcascade_frontalface_alt2.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.3, minNeighbors = 7);
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (-1, 512, -1), 50)
```

```python
plt.imshow(convertToRGB(test_image))
```
